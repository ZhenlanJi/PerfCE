```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from os.path import dirname
# import seaborn as sns
# sns.set_theme()
```

```python
test_df = pd.read_csv(os.path.join(
    dirname(dirname(os.getcwd())), 'inference', 'mysql', 'test_1m.csv'))
# test_df=(test_df) / (test_df.max() - test_df.min())
data = test_df["Current_QPS__None"].to_numpy()
# data=(data-data.min())/(data.max()-data.min())
params = stats.norm.fit(data, method='mle')
norm = stats.norm(*params)
kde = stats.gaussian_kde(data)

# print(distribution.mean())
print(data.mean())

# fig, ax = plt.subplots(1, 1)
npts_sample = int(1e3)
x = np.linspace(data.min(), data.max(), npts_sample)
kde_pdf = kde.evaluate(x)
norm_pdf=norm.pdf(x)
plt.hist(data, density=True, bins=50, label='Real Frequency', alpha=0.8, color='#1f78b4', zorder=1)
plt.plot(x, kde_pdf, label='Gaussian KDE', color="#fb9a99", zorder=2)
plt.plot(x, norm_pdf, label='Normal', color="#b2df8a", zorder=3)

plt.fill_between(x, 0, kde_pdf, alpha=0.3, color="#fb9a99", zorder=2)
plt.fill_between(x, 0, norm_pdf, alpha=0.3, color="#b2df8a", zorder=3)
plt.legend()
plt.rcParams['figure.figsize'] = (11,5) 
plt.xlim(0, 6000)
plt.xlabel('Current QPS')
plt.ylabel('Probability Density')
# set font size
plt.rcParams.update({'font.size': 15})

# plt.show()
# save figure as pdf
# plt.savefig('distribution.pdf')


# plt.hist(y, density=True, bins=50)

# plt.plot(list(range(10000)),y)
# plt.plot(y, density=True, bins=50)
# plt.show()
```
